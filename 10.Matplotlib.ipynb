# 1.导入所需要的包

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
```

# 2.创建我们的数据 
## names = ['5', '10', '15', '20', '25']
## x = range(len(names))
## y = [0.855, 0.84, 0.835, 0.815, 0.81]
## y1=[0.86,0.85,0.853,0.849,0.83]

```python
names = ['5', '10', '15', '20', '25']
x = range(len(names))
y = [0.855, 0.84, 0.835, 0.815, 0.81]
y1=[0.86,0.85,0.853,0.849,0.83]
```

# 3.防止乱码 设置中文 设置为黑体

```python
my_font = font_manager.FontProperties(fname='../data/STSONG.TTF')
```

# 4.设置两条曲线x,y和x,y1 图例分别为曲线图1和曲线图2
# 5. 显示图例
# 6.设置x轴的刻度疏密程度为x值，标签值为names，45位旋转的度数
# 7.设置x轴的名称和y轴的名称

```python
plt.plot(x, y,label='曲线图1')
plt.plot(x, y1,label='曲线图2')
plt.legend(prop=my_font) # 让图例生效
plt.xticks(x, names, rotation=45)
plt.xlabel("时间", fontproperties=my_font) #X轴标签
plt.ylabel("刻度值", fontproperties=my_font) #Y轴标签
plt.title("时间和刻度的变化", fontproperties=my_font) #标题
 
plt.show()
```
